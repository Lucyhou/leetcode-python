# 579 Find Cumulative Salary of an Employee

```python
import pandas as pd
import re
import numpy as np
Employee= pd.DataFrame({'id':[1,2,1,2,3,1,3,1,3],'Month':[1,1,2,2,2,3,3,4,4],'Salary':[20,20,30,30,40,40,60,60,70]})
Employee=Employee[['id','Month','Salary']]
Employee
```

```python
Employee.groupby(['id','Month'])['Salary'].idxmax()
```

```python
output=Employee.loc[Employee.groupby(['id','Month'])['Salary'].idxmax()].sort_values(by=['id','Month'],ascending=[True,False])
output
```

```python
output2=output.groupby(['id'], sort=False)['Month'].max().to_frame()
output3=output2.reset_index()
output3
```

```python
outp=pd.merge(output,output3,on=['id','Month'],how='inner')
outp
```

```python
outpp=pd.merge(output,outp,on=['id','Month'],how='left')
outpp
```

```python
newtable=outpp[outpp.Salary_y.isnull()][['id','Month','Salary_x']]
newtable
```

```python
new1=newtable.groupby(by=['id','Month']).sum().groupby(level=[0]).cumsum().reset_index()
new1
```

```python
new1.sort_values(by=['id','Month'],ascending=[True,False])
```

```python
print(new1.sort_values(by=['id','Month'],ascending=[True,False]).to_string(index=False))
```
