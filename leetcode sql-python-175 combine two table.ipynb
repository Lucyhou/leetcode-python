```python
import pandas as pd
Person = pd.DataFrame({'PersonId':[1,2,3],
                       'FirstName':['Jonnan','Luke','Joe'],
                       'LastName':['Bach','Doe','Lee']})
Address = pd.DataFrame({'AddressId':[1,2],
                        'PersonId':[2,3],
                        'City':['New York','Hayward'],
                        'State':['NY','CA']})
```

```python
Person
```

```python
Address
```

```python
a = pd.merge(Person,Address,how = 'left',on = 'PersonId')
a

```
